```python
import scipy.stats as stat
```

```python
#Box-cox transformation for DMC column
sns.boxplot(agn_data['DMC'])
plt.title('Box plot before outlier removing')
plt.show()
agn_data['DMC'],fitted_lambda=stat.boxcox(agn_data['DMC'],lmbda=None)
sns.boxplot(agn_data['DMC'])
plt.title('Box plot after outlier removing')
plt.show()

```

```python
#Box-cox transformation for FFMC column
sns.boxplot(agn_data['FFMC'])
plt.title('Box plot before outlier removing')
plt.show()
agn_data['FFMC'],fitted_lambda=stat.boxcox(agn_data['FFMC'],lmbda=None)
sns.boxplot(agn_data['FFMC'])
plt.title('Box plot after outlier removing')
plt.show()
```

```python
#yeojohnson transformation for Rain column
sns.boxplot(agn_data['Rain'])
plt.title('Box plot before outlier removing')
plt.show()
agn_data['Rain'],lmbda=stat.yeojohnson(agn_data['Rain'])
sns.boxplot(agn_data['Rain'])
plt.title('Box plot after outlier removing')
plt.show()
```

```python
#Mean transform for Ws column
sns.boxplot(agn_data['Ws'])
plt.title('Box plot before outlier removing')
plt.show()
for i in agn_data['Ws']:
    q1=agn_data['Ws'].quantile(0.25)
    q3=agn_data['Ws'].quantile(0.75)
    iqr=q3-q1
    Lower_tail=q1-(1.5*iqr)
    Upper_tail=q3+(1.5*iqr)
    
    if i>Upper_tail or i<Lower_tail:
        agn_data['Ws']=agn_data['Ws'].replace(i,np.mean(agn_data['Ws']))
        
sns.boxplot(agn_data['Ws'])
plt.title('Box plot after outlier removing')
plt.show()        
```

```python
#Finding outliers after transformation
plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in agn_data:
    if plotnumber<=9 :
        ax = plt.subplot(3,3,plotnumber)
        sns.boxplot(agn_data[column])
        plt.xlabel(column,fontsize=20)
        #plt.ylabel('Salary',fontsize=20)
    plotnumber+=1
plt.show()
```

```python
Precision = true_positive/(true_positive+false_positive)
Precision
```

```python
Recall = true_positive/(true_positive+false_negative)
Recall
```
